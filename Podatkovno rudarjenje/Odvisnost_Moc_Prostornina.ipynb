```python
import matplotlib.pyplot as plt
import statistics
from sklearn.linear_model import LinearRegression
```

```python
import operator
import numpy as np
import os
from csv import DictReader

datoteke = os.listdir("podatki/")
podatki_o_vozilih = []
for dat in datoteke:
    if "Podatki" in dat and "2020" not in dat:
        neki = DictReader(open('podatki/' + dat, 'rt', encoding='ANSI'), delimiter=';')
        for i in neki:
            if i["P.1.2-Nazivna moc"] != "" and i["P.1.2-Nazivna moc"] != None and i["P.1.2-Nazivna moc"] != "Bencin" and i["P.1.2-Nazivna moc"] != "-1.0":
                podatki_o_vozilih.append([float(str(i["P.1.2-Nazivna moc"]).replace(",", ".")), float(str(i["P.1.1-Delovna prostornina"]).replace(",", ".")), i["J-Kategorija in vrsta vozila (opis)"]])
podatki_o_vozilih = np.array(podatki_o_vozilih)
```

```python
avto = podatki_o_vozilih[podatki_o_vozilih[:,2] == "osebni avtomobil"]
```

```python
motor = podatki_o_vozilih[podatki_o_vozilih[:,2] == "motorno kolo"]
```

```python
moped = podatki_o_vozilih[podatki_o_vozilih[:,2] == "moped-dvokolo"]
```

```python
stirikolesnik = podatki_o_vozilih[podatki_o_vozilih[:,2] == "štirikolo"]
```

```python
tovorno = podatki_o_vozilih[podatki_o_vozilih[:,2] == "tovorno vozilo"]
```

```python
vlecno = podatki_o_vozilih[podatki_o_vozilih[:,2] == "vlečno vozilo"]
```

```python
def prikazi_graf(x, y, tip, mi, ma):
    avti = dict()
    for i in range(len(x)):
        if x[i] not in avti:
            avti[float(x[i])] = []
        avti[float(x[i])].append(float(y[i]))
    for i in avti:
        avti[float(i)] = statistics.mean(avti[i])
    slovar = dict()
    for key in sorted(avti):
        slovar[key] = avti[key]
    x2 = [float(i) for i, j in slovar.items()]
    y2 = [float(j) for i, j in slovar.items()]
    plt.plot(x2, y2)
    plt.xlabel("Prostornina motorja ccm")
    plt.ylabel("Moč motorja kW")
    plt.title("Gibanje moči glede na prostornino motorja .." + tip + "..")
    plt.show()

    model = LinearRegression(fit_intercept=True)
    x2 = np.array(x2)
    y2 = np.array(y2)
    model.fit(x2[:, np.newaxis], y2)
    xfit = np.linspace(mi, ma, 1000)
    yfit = model.predict(xfit[:, np.newaxis])
    plt.xlabel("Prostornina motorja ccm")
    plt.ylabel("Moč motorja kW")
    plt.title("Premica linearne regresije .." + tip + "..")
    plt.plot(x2, y2)
    plt.plot(xfit, yfit, color="red")
    plt.show()
```

```python
x = avto[:,1].astype("float")
y = avto[:,0].astype("float")
prikazi_graf(x,y,"osebni avtomobil", min(x), max(x))
x = motor[:,1].astype("float")
y = motor[:,0].astype("float")
prikazi_graf(x,y,"motor", min(x), max(x))
x = moped[:,1].astype("float")
y = moped[:,0].astype("float")
prikazi_graf(x,y,"moped", min(x), max(x))
x = vlecno[:,1].astype("float")
y = vlecno[:,0].astype("float")
prikazi_graf(x,y,"vlečno vozilo", min(x), max(x))
x = tovorno[:,1].astype("float")
y = tovorno[:,0].astype("float")
prikazi_graf(x,y,"tovorno vozilo", min(x), max(x))
x = stirikolesnik[:,1].astype("float")
y = stirikolesnik[:,0].astype("float")
prikazi_graf(x,y,"štirikolesnik", min(x), max(x))
```

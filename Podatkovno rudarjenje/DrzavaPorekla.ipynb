```python
import operator
import numpy as np
import os
from csv import DictReader
import matplotlib.pyplot as plt


datoteke = os.listdir("podatki/")
obcine = []
for dat in datoteke:
    if "Podatki" in dat and "2020" not in dat:
        neki = DictReader(open('podatki/' + dat, 'rt', encoding='ANSI'), delimiter=';')
        for i in neki:
            obcine.append([i["D.4.2-Drzava (opis)"], i["Status vozila (opis)"]])
obcine1 = np.array(obcine)
```

```python
stevilo_na_drzavo = dict()
for i in obcine1[obcine1[:,0] != "X"]:
    if i[1] != "odjavljeno" and i[0] != None:
        if i[0] not in stevilo_na_drzavo:
            stevilo_na_drzavo[i[0]] = 0
        stevilo_na_drzavo[i[0]] += 1
```

```python
stevilo_na_drzavo = sorted(stevilo_na_drzavo.items(),key=operator.itemgetter(1),reverse=True)
```

```python
print("Na novo registrirana vozila so iz", len(stevilo_na_drzavo), "držav.")
```

```python
stevilo_na_drzavo = stevilo_na_drzavo[:20]
```

```python
plt.xlabel("Število avtomobilov")
plt.ylabel("Država")
plt.xscale("linear")
plt.title('Število avtomobilov na državo porekla')
plt.barh([i[0] for i in stevilo_na_drzavo], [i[1] for i in stevilo_na_drzavo])
plt.show()
```

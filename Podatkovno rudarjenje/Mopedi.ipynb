```python
from csv import DictReader
import operator
import matplotlib.pyplot as plt

mopedi = dict()

for j in range(2015,2020):    
    for i in range(1,13):
        file = 'podatki/Podatki_{:02d}'.format(i) + str(j) + '.csv'
        podatki = DictReader(open(file, 'rt', encoding='ANSI'), delimiter=';')
        
        if str(i) + '_' + str(j) not in mopedi:
            mopedi[str(i) + '_' + str(j)] = 0

        for row in podatki:
            status = str(row['Status vozila (opis)'])
            try:
                vrsta_vozila = str(row['J-Kategorija in vrsta vozila (opis)'])
                
                if status == 'registrirano' and vrsta_vozila == 'moped-dvokolo':
                    mopedi[str(i) + '_' + str(j)] += 1
                    
            except:
                #Napaka v datumu vozila
                None
```

```python
meseci = list(mopedi.keys())
st_mopedov = list(mopedi.values())

plt.figure(figsize=(20,10))
plt.plot(meseci, st_mopedov, label='Registriranih mopedov')
plt.xlabel('Meseci')
plt.xticks(rotation=45)
plt.ylabel('Število registriranih mopedov')
plt.legend()
plt.show()
```
